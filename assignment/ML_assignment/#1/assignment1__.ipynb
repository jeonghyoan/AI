#### 1. Load the dataset

```python
import pandas as pd
import numpy as np

#data load
data=pd.read_csv(r'C:\Users\hyoan\Desktop\data.csv', sep=',')

data
```

#### 2. Split the dataset (Training / Validation)

```python
data.drop(["id","Unnamed: 32"],axis=1,inplace=True)

data.diagnosis=[1 if each == "M" else 0 for each in data.diagnosis]

data
```

```python
from sklearn.model_selection import train_test_split

X_data=data.drop(["diagnosis"], axis=1)

#Nomalization
X=(X_data-np.min(X_data))/(np.max(X_data) -np.min(X_data))

y=data['diagnosis'].values


#data split
X_train, X_val, y_train, y_val = train_test_split(X, y, stratify=y, random_state=42)
```

#### 3. Train the models while changing some hyperparameters
#### 5. Performance tables for training set / validation set according to hyperparameter settings

→Learning algorithms used (hyperparameter)

1.	Decision Tree (min_samples_leaf, max_depth)

2.	Logistic Regression (C)


- Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

train_accuracy = []
val_accuracy = []

msl_settings = [1, 2, 5, 7, 10, 20] # code will iterate over these six settings with the for statement

for msl in msl_settings:
        # model training with labeled training data
        clf = DecisionTreeClassifier(min_samples_leaf= msl, random_state=0) # it repeats by changing only this msl
        clf.fit(X_train, y_train)

        # prediction
        y_train_hat = clf.predict(X_train)
        y_val_hat = clf.predict(X_val)

        # evaluation. / in every iteration, store the results. 
        train_accuracy.append(accuracy_score(y_train, y_train_hat))
        val_accuracy.append(accuracy_score(y_val, y_val_hat))
        

        
train_accuracy2 = []
val_accuracy2 = [] 

mxd_settings = [2,3,4,5,6,7,8]

for mxd in mxd_settings:
        # model training with labeled training data
        clf2 = DecisionTreeClassifier(max_depth = mxd, random_state=0) # it repeats by changing only this mxd
        clf2.fit(X_train, y_train)

        # prediction
        y_train_hat2 = clf2.predict(X_train)
        y_val_hat2 = clf2.predict(X_val)

        # evaluation. / in every iteration, store the results. 
        train_accuracy2.append(accuracy_score(y_train, y_train_hat2))
        val_accuracy2.append(accuracy_score(y_val, y_val_hat2))
```

```python
# check the result
result = pd.DataFrame({"min_samples_leaf":msl_settings, "training accuracy": train_accuracy, "validation accuracy": val_accuracy})
result2 = pd.DataFrame({"max_depth":mxd_settings, "training accuracy": train_accuracy2, "validation accuracy": val_accuracy2})
```

- Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

train_accuracy3 = []
val_accuracy3 = []

C_settings = [0.01, 0.1, 1, 10, 100, 1000, 10000]

for C in C_settings:
    logreg = LogisticRegression(C=C)
    logreg.fit(X_train, y_train)
    
    y_train_hat3 = logreg.predict(X_train)
    y_val_hat3 = logreg.predict(X_val)
    
    train_accuracy3.append(accuracy_score(y_train, y_train_hat3))
    val_accuracy3.append(accuracy_score(y_val, y_val_hat3))
```

```python
pd.DataFrame({"C":C_settings, "training accuracy": train_accuracy3, "validation accuracy": val_accuracy3})
```
